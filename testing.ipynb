```python


n = int(input("nhap mot so bat ki: "))

if n < 2:
    print("khong phai la so nguyen to")
else:
    # Sửa lỗi 1: n**0.5 dùng dấu chấm, không dùng dấu phẩy
    # Sửa lỗi 2: range phải bọc trong int()
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            print(n, "khong phai la so nguyen to") # Sửa logic: Chia hết => KHÔNG phải
            break # Thoát ngay lập tức
    else:
        # Đây là cấu trúc đặc biệt của Python:
        # Câu lệnh else này chỉ chạy khi vòng lặp chạy hết MÀ KHÔNG GẶP break
        print("la so nguyen to:", n)
```
