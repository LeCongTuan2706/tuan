```python
# Nhập hai số nguyên a và b
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))

# Tính thương và phần dư khi chia a cho b
thuong = a // b     # phép chia lấy phần nguyên
phan_du = a % b     # phép chia lấy phần dư

# In kết quả ra màn hình
print("Thương:", thuong)
print("Phần dư:", phan_du)
```

```python
# Nhập 6 điểm số nguyên
a1 = int(input("Nhập điểm 1: "))
b1 = int(input("Nhập điểm 2: "))
c1 = int(input("Nhập điểm 3: "))
a2 = int(input("Nhập điểm 4: "))
b2 = int(input("Nhập điểm 5: "))
c2 = int(input("Nhập điểm 6: "))

# Tính điểm trung bình theo công thức: TB = ((a1+b1+c1)+(a2+b2+c2)*2)/10
TB = ((a1 + b1 + c1) + (a2 + b2 + c2) * 2) / 10

# In kết quả, làm tròn 1 chữ số thập phân
print("Điểm trung bình:", round(TB, 1))
```
